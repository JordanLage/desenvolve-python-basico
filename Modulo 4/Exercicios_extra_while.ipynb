### 1. Faça um programa, utilizando while e listas, que permita o usuário escrever o nome de cinco pessoas e os mostre na tela.


```python
lista = []
cont = 0

while cont < 5:
    lista.append(input('Digite um nome: '))
    cont += 1
print(f'Nomes: {lista}')
```

### 2. Faça um programa, utilizando while, que permita o usuário fazer contas de adição enquanto quiser.



```python
repetir = 'S'
while repetir == 'S':
    n1 = int(input('Digite um número: '))
    n2 = int(input('Digite outro número: '))
    soma = n1 + n2
    print(f'{n1} + {n2} = {soma}')
    repetir = str(input('Deseja realizar mais uma soma? [S ou N]'))
```

### 3. Faça um programa, utilizando while e listas, que permita o usuário realizar o cadastro de um número indeterminado de pessoas enquanto quiser e os mostre na tela ao finalizar.


```python
lista = []
print('Digite 0 para terminar o cadastro!')

while True:
    nome = str(input('Digite o nome do funcionário: '))
    if nome == '0':
        break
    else:
        lista.append(nome)
    
print(f'Funcionários: {lista}')
```
